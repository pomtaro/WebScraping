```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.google.com/search?q=Beautifulsoup&rlz=1C5CHFA_enJP823JP824&oq=Beautifulsoup&aqs=chrome..69i57j0l2j69i59l2j0.4419j0j8&sourceid=chrome&ie=UTF-8'
r = requests.get(url)

html_soup = BeautifulSoup(r.content, 'html.parser')
```

```python
"""
> を使うことで、親子関係を表現できる
使い方がいまいちわからない・・・
"""

links_tags = html_soup.select('a')
links_tags_filtered = html_soup.select('p > a')
```

```python
print(len(links_tags))
print(len(links_tags_filtered))

print('links_tags')

print(links_tags)

print('links_tags_filtered')

print(links_tags_filtered)
```
